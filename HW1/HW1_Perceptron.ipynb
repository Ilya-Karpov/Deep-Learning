```python
import torch
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from torch.utils.data import TensorDataset, DataLoader
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
data = pd.read_csv("housing.csv", header = None, delimiter = r"\s+", names = column_names)
```

```python
data.head()
```

```python
data.info()
```

```python
scaler = preprocessing.MinMaxScaler()
names = data.columns
d = scaler.fit_transform(data)
df = pd.DataFrame(d, columns=names)
df.head()
```

```python
features = df.drop('MEDV', axis=1).to_numpy()
target = df.MEDV.to_numpy()
```

```python
features = torch.tensor(features, dtype = torch.float32)
target = torch.tensor(target, dtype = torch.float32)
#X = X.clone().detach().requires_grad_(True)
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
batch_size = 10
dataset = TensorDataset(features, target)
# Randomly reading mini-batches
data_iter = DataLoader(dataset, batch_size, shuffle=True)

# Read a batch to see how it works
for X, y in data_iter:
    print(X, y)
    break
```

```python
model = torch.nn.Sequential(torch.nn.Linear(13, 1))
```

```python
loss = torch.nn.MSELoss(reduction='mean')
```

```python
#trainer = torch.optim.SGD(model.parameters(), lr=0.001)
```

```python
trainer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 1000
for epoch in range(1, num_epochs + 1):
    for X, y in data_iter:
        trainer.zero_grad()
        l = loss(model.forward(X).reshape(-1), y)
        l.backward()
        trainer.step()
    l = loss(model.forward(features).reshape(-1), target)
    if epoch % 100 == 0:
        print('epoch %d, loss %f' % (epoch, l.item()))
        print('w',model[0].weight.data)
        print('b',model[0].bias.data)
```
